```python
import pandas as pd
import datetime as dt
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
```

```python
data = pd.read_csv('Dados_PRF_2023.csv', encoding = 'latin1', sep = ';')
```

```python
data['data_inversa'] = pd.to_datetime(data['data_inversa'])

monthlyIncidents2023 = data.groupby(data['data_inversa'].dt.to_period('M')).size().reset_index(name='incidents')

monthlyIncidents2023['month'] = monthlyIncidents2023['data_inversa'].dt.strftime('%Y-%m')

fig = px.bar(monthlyIncidents2023, x='month', y='incidents', labels={
                     "month": "Mês",
                     "incidents": "Acidentes"
                 }, title="Números de Acidentes por Mês", height=500)

fig.update_traces(texttemplate='%{y}', textposition='outside')

fig.update_yaxes(showticklabels=False)

fig.show()
```

```python
weeklyIncidents2023 = data['dia_semana'].value_counts().reset_index()
weeklyIncidents2023.columns = ['day_of_week', 'incidents']

desiredOrder = ['segunda-feira', 'terça-feira', 'quarta-feira', 'quinta-feira', 'sexta-feira', 'sábado', 'domingo']

weeklyIncidents2023['day_of_week'] = pd.Categorical(weeklyIncidents2023['day_of_week'], categories=desiredOrder, ordered=True)

weeklyIncidents2023 = weeklyIncidents2023.sort_values('day_of_week')

fig = px.bar(weeklyIncidents2023, x='day_of_week', y='incidents', labels={
                     "day_of_week": "Dia da Semana",
                     "incidents": "Acidentes"
                 }, title="Números de Acidentes por Dias da Semana", height=500)

fig.update_traces(texttemplate='%{y}', textposition='outside')

fig.update_yaxes(showticklabels=False)

fig.show()
```

```python
data['horario'] = pd.to_datetime(data['horario'])

hourlyIncidents2023 = data.groupby(data['horario'].dt.hour).size().reset_index(name='incidents')

hourlyIncidents2023.rename(columns={'horario' : 'hour'}, inplace=True)

fig = px.bar(hourlyIncidents2023, x='hour', y='incidents', labels={
                     "hour": "Horário",
                     "incidents": "Acidentes"
                 }, title="Números de Acidentes por Hora", height=500)
fig.update_traces(texttemplate='%{y}', textposition='outside')
fig.update_layout(
    xaxis=dict(
        tickmode='array',
        tickvals=list(range(24)),
        ticktext=[f'{hour:02d}:00' for hour in range(24)]
    )
)
fig.show()
```

```python
statesIncidents2023 = data['uf'].value_counts().reset_index()
statesIncidents2023.columns = ['states', 'incidents']

statesIncidents2023 = statesIncidents2023.sort_values('states')

fig = px.bar(statesIncidents2023, x='states', y='incidents', labels={
                     "states": "Estados",
                     "incidents": "Acidentes"
                 }, title="Números de Acidentes por Estados", height=500)

fig.update_traces(texttemplate='%{y}', textposition='outside')

fig.update_yaxes(showticklabels=False)

fig.show()
```

```python
data['city'] = data['municipio'] + ' - ' + data['uf']

citiesIncidents2023 = data['city'].value_counts().reset_index()
citiesIncidents2023.columns = ['city', 'incidents']

citiesIncidents2023 = citiesIncidents2023.sort_values('city')

top20CitiesIncidents2023 = citiesIncidents2023.nlargest(20, 'incidents')

fig = px.bar(top20CitiesIncidents2023, x='city', y='incidents', labels={
                     "city": "Cidades",
                     "incidents": "Acidentes"
                 }, title="Números de Acidentes por Top 20 Cidades", height=600)

fig.update_traces(texttemplate='%{y}', textposition='outside')

fig.update_yaxes(showticklabels=False)

fig.show()
```

```python
data['highway'] = data['br'].astype(str) + ' - ' + data['uf']

highwaysIncidents2023 = data['highway'].value_counts().reset_index()
highwaysIncidents2023.columns = ['highway', 'incidents']

highwaysIncidents2023 = highwaysIncidents2023.sort_values('highway')

top20HighwaysIncidents2023 = highwaysIncidents2023.nlargest(20, 'incidents')

fig = px.bar(top20HighwaysIncidents2023, x='highway', y='incidents', labels={
                     "highway": "Estradas",
                     "incidents": "Acidentes"
                 }, title="Números de Acidentes por Top 20 Estradas", height=600)

fig.update_traces(texttemplate='%{y}', textposition='outside')

fig.update_yaxes(showticklabels=False)

fig.show()
```

```python
from random import randrange

data['latitude'] = data['latitude'].apply(lambda x: x.replace(',', '.')).astype('float')
data['longitude'] = data['longitude'].apply(lambda x: x.replace(',', '.')).astype('float')
data.head()
```

```python
aaa = data['uf'].value_counts().reset_index()
aaa.columns = ['uf2', 'acidentes']
data2 = data
data2 = data2.drop(columns='acidentesQtd')
data2 = data2.merge(aaa, left_on='uf', right_on='uf2')
data2 = data2.drop(columns='uf2')
data2.head()
```

```python
# Group and count cities incidents

data['location'] = data['latitude'].astype(str) + ';' + data['longitude'].astype(str)
localizationIncidents2023 = data['location'].value_counts().reset_index()

fig = go.Figure()

scale = 5000
r = randrange(255)
g = randrange(255)
b = randrange(255)
rand_color = (r, g, b)

styles= ['open-street-map', 'white-bg', 'carto-positron',
         'carto-darkmatter', 'stamen-terrain', 'stamen-toner']

px.scatter_mapbox(data,
                  lat = data['latitude'],
                  lon = data['longitude'],
                  mapbox_style=styles[0],
                  width = 1200,
                  height = 600,
                  zoom = 4,
                  )
```

```python
haveDeadData = data2.where(data2['mortos'] > 0)

statesDeath2023 = haveDeadData['uf'].value_counts().reset_index()
statesDeath2023.columns = ['state', 'incidents']

fig = px.bar(statesDeath2023, x='state', y='incidents', labels={
                     "uf": "Estradas",
                     "incidents": "Mortos"
                 }, title="Números de Mortes por Estados", height=600)

fig.update_traces(texttemplate='%{y}', textposition='outside')

fig.update_yaxes(showticklabels=False)

fig.show()
```
